```python
import numpy as np
import pandas as pd
import datetime
import os
import time
```

# 导入4.27数据

```python
date = '4.27'
# 跨中风速
path_wind = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中风速统计值.csv".format(date)

# 跨中竖向加速度
path_Accl_z = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中竖向加速度统计值.csv".format(date)

# 跨中横向加速度
path_Accl_x = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中横向加速度统计值.csv".format(date)

# 跨中竖向位移
path_Disp_z = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中竖向位移统计值.csv".format(date)

# 跨中水平位移
path_Disp_xy = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中水平位移统计值.csv".format(date)

# 跨中温湿度
path_T = r"..\数据\浦仪夹江大桥数据{0}\processed_data\跨中温湿度.csv".format(date)
```

```python
def readFile(path):
    return pd.read_csv(path,header=0)
df_wind = readFile(path_wind)
df_Accl_z = readFile(path_Accl_z)
df_Accl_x = readFile(path_Accl_x)
df_Disp_z = readFile(path_Disp_z)
df_Disp_xy = readFile(path_Disp_xy)
df_T = readFile(path_T)

# 将数据合并到一张表
df = pd.concat([df_wind, df_T, df_Accl_z, df_Accl_x, df_Disp_xy, df_Disp_z], axis = 1, ignore_index = False, join = "inner")
df = df[["采集时段",
    "总风速","纵桥向风速","横桥向风速","竖向风速","Ix","Iy","Iz",
         "温度平均值", "湿度平均值",
    "纵向变位平均值","纵向变位标准差","横向变位平均值","横向变位标准差","竖向变位平均值","竖向变位标准差",
   "横向加速度平均值","横向加速度标准差",
   "竖向加速度平均值","竖向加速度标准差"
]]
df.to_csv(r"..\数据\浦仪夹江大桥数据4.27\processed_data\final.csv")
```
