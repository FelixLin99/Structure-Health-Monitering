# 自助式作图模块
##### $Tutorial：$
   - 作图函数：midSpanData.show(x, y)
   - 可画时域图和相关性分析图，相关参数说明如下
    </br></br>
       - 时域图
           - x = ' $time$ '
           - y = $attr$  或  $list (attr)$
           </br></br>
       - 相关性分析图
           - x = $attr$
           - y = $attr$ 或 $list (attr)$
           </br></br>
   - attr 可选如下：
       - 风属性：
           - '总风速', '纵桥向风速', '横桥向风速', '竖向风速', 'Ix', 'Iy', 'Iz'
           </br></br>
       - 温湿度属性：
           - '温度平均值','湿度平均值'
           </br></br>
       - 桥响应属性：
           - '纵向变位平均值', '纵向变位标准差', '横向变位平均值', '横向变位标准差', '竖向变位平均值','竖向变位标准差', '横向加速度平均值', '横向加速度标准差', '竖向加速度平均值', '竖向加速度标准差'



```python
from Visualization import MidSpanData
midSpanData = MidSpanData(date='4.27')  # - data = '4.27' 或 '5.4'
```

```python
# 做时域图
x = 'time'
y = '总风速'
midSpanData.show(x, y)
```

```python
# 做时域图
x = 'time'
y = ['Ix', 'Iy']
midSpanData.show(x, y)
```

```python
# 做关联分析图
x = '总风速'
y = ['横向加速度标准差', '竖向加速度标准差']
midSpanData.show(x, y)
```
